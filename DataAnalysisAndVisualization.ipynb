```python
#Dataset Creation
import pandas as pd
import numpy as np
n = 1000 #try changing this
data = {
'ProductID': np.arange(1, n + 1),
'ProductName': ['Product_' + str(i) for i in range(1, n + 1)],
'Category': np.random.choice(['Electronics', 'Clothing', 'Home Decor', 'Books'], size=n),
'Brand': np.random.choice(['Brand_A', 'Brand_B', 'Brand_C', 'Brand_D'], size=n),
'Price': np.round(np.random.uniform(10, 200, size=n), 2),
'DiscountPercentage': np.round(np.random.uniform(5, 30, size=n), 2),
'StockAvailability': np.random.randint(1, 100, size=n),
'Rating': np.round(np.random.uniform(3, 5, size=n), 2),
'Offer': np.random.choice(['Buy One Get One', 'Flat 20% Off', 'Free Shipping', 'No Offer'], size=n),
'ShippingCost': np.round(np.random.uniform(0, 10, size=n), 2)
}
df = pd.DataFrame(data)
print(df.head(10)) #by default --> 5
```

```python
#CASE STUDY
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Selecting relevant features
features = ['DiscountPercentage', 'StockAvailability', 'ShippingCost']
target = 'Price'

# Creating feature matrix X and target variable y
X = df[features]
y = df[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build the linear regression model
model = LinearRegression() #try fine tuning the algorithm
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Visualize the predicted vs. actual prices
plt.scatter(X_test['DiscountPercentage'], y_test, color='brown', label='Actual Prices')
plt.scatter(X_test['DiscountPercentage'], y_pred, color='green', label='Predicted Prices')
plt.xlabel('Discount Percentage')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Function to recommend an optimal price for a user
def get_optimal_price(model, user_input_dict):
    user_df = pd.DataFrame([user_input_dict])
    optimal_price = model.predict(user_df)[0]
    return optimal_price

discount = float(input("Enter the Discount %: "))
stock = float(input("Enter the Stock Availability: "))
shippingCost = float(input("Enter the Shipping Cost: "))

user_data = {
    'DiscountPercentage': discount,
    'StockAvailability': stock,
    'ShippingCost': shippingCost
}


optimal_price = get_optimal_price(model, user_data)
print("Optimal Price Recommendation: â‚¹", round(optimal_price, 2))
```
